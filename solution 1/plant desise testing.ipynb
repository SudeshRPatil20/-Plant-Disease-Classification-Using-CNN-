```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
ImageDataGenerator=tf.keras.preprocessing.image.ImageDataGenerator()
```

```python
model = tf.keras.models.load_model("D:/sudesh/projects/plant desies detection/archive (4)/train_model.keras")
```

```python
model.summary()
```

```python
import cv2
```

```python
image_path = "D:/sudesh/projects/plant desies detection/archive (4)/Test/Test/Healthy/8ddaa5a5caa5caa8.jpg"
```

```python
img = cv2.imread(image_path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("test image")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
image= tf.keras.preprocessing.image.load_img(image_path,target_size=(256, 256))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
inpur_arr = np.array([input_arr]) #convert image to array
print(inpur_arr.shape)
```

```python
prediction= model.predict(inpur_arr)
prediction,prediction.shape
```

```python
result_index = np.argmax(prediction)
result_index
```

```python
class_name= ['Healthy', 'Powdery', 'Rust']
```

```python
# model_prediction
model_prediction = class_name[result_index]
plt.imshow(img)
plt.title(f"plant leaf is : {model_prediction}")
plt.xticks([])
plt.yticks([])
plt.show()
```
